# One Hot Encoding 

```python
import pandas as pd 
```

```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Loan%20Eligibility%20Prediction.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
encoded_data = df[['Gender','Married']]
encoded_data
```

```python
# Using get dummies 
pd.get_dummies(encoded_data).info()
```

```python
# using sklearn library 
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ar=ohe.fit_transform(encoded_data).toarray()
```

```python
pd.DataFrame(ar,columns=['Gender_Female','Gender_Male','Married_No','Married_Yes'])
```
